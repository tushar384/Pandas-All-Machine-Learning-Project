```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
def preprocessing_inputs(df,target='Price'):
    df = df.copy()
    missing_target_rows = df.loc[df['Per_Sqft'].isna(),:].index
    df = df.drop(missing_target_rows,axis = 0).reset_index(drop =True)
    
    # fill na value 
    df['Parking'] =df['Parking'].fillna(df['Parking'].mean())
    df['Bathroom'] =df['Bathroom'].fillna(df.Bathroom.mean())
    #df['Per_Sqft'] = df['Per_Sqft'].fillna(df['Per_Sqft'].mean())
    # fill na value binary object columns
    for column in ['Furnishing','Type']:
        df[column] = df[column].fillna(df[column].mode()[0])
    # binary encoding columns
   
   
  
    # one hot encode
    df = pd.get_dummies(df,drop_first = True)
    
    # split data 
    if target=='Per_Sqft':
        df = df.drop(['Price'],axis = 'columns')
        y = df['Per_Sqft'] 
        X = df.drop(['Per_Sqft'],axis = 'columns')
    elif target=='Price':
        df = df.drop(['Per_Sqft'],axis = 'columns')
        y = df['Price']
        X = df.drop(['Price'],axis = 'columns')
    # train test split
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,shuffle =True,random_state = 1)
    # scaling data
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train),index = X_train.index,columns = X_train.columns)
    X_test  = pd.DataFrame(scaler.transform(X_test),index = X_test.index,columns = X_test.columns)
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test =preprocessing_inputs(data,target = 'Price')
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression,Ridge,Lasso
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC,LinearSVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'Logistic_Rgression'   : LogisticRegression(),
    'Linear_Rgression'     : LinearRegression(),
    'Ridge'                : Ridge(),
    'Lasso'                : Lasso(),
    'svc'                  : SVC(),
    'Linear_svc'           : LinearSVC(),
    'Decision_tree'        : DecisionTreeClassifier(),
    'Random_forest'        : RandomForestClassifier(),
    'Gradient_Boosting'    : GradientBoostingClassifier(),
    'Neural_Network'       : MLPClassifier(),
    'K_Neighbors'          : KNeighborsClassifier(),
     'GNB'                 : GaussianNB()
}
for name, model in models.items():
    model.fit(X_train,y_train)
    print(name + 'Trannied')
```

```python
for name, model in models.items():
    print(name + ' Accuracy : {:2f}'.format(model.score(X_test,y_test)))
```
