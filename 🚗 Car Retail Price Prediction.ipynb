```python
def preprocessing_inputs(df):
    df = df.copy()
    # fill na value
    for column in ['Engine HP','Engine Cylinders'] :
        df[column] = df[column].fillna(df[column].mean())
    # fillna value catagorical column
    df['Engine Fuel Type'] = df['Engine Fuel Type'].fillna(df['Engine Fuel Type'].mode()[0])
   
    df['Number of Doors'] = df['Number of Doors'].fillna(df['Number of Doors'].mean())
    # # fillna catagorical column
    df['Market Category'] = df['Market Category'].fillna(df['Market Category'].mode()[0])
    # data get dummies
    df = pd.get_dummies(df,drop_first = True)
    # split df in X,y
    y = df['MSRP']
    X = df.drop(['MSRP'],axis = 'columns')
    # train model 
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test = preprocessing_inputs(data)
```
