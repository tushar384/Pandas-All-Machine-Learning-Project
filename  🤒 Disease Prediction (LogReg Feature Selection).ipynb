```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
def preprocessing_inputs(df):
    df = df.copy()
    # too many value in is NAn value remove those columns
    df = df.drop(['Unnamed: 133'],axis = 'columns')
    # split df into X,y
    y = df['prognosis']
    X = df.drop(['prognosis'],axis = 'columns')
    # train data sets 
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
   # scaler = fit(X_train) 
    X_train =pd.DataFrame(scaler.fit_transform(X_train),index = X_train.index,columns = X_train.columns)
    X_test = pd.DataFrame(scaler.fit_transform(X_test),index = X_test.index,columns = X_test.columns)
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test = preprocessing_inputs(train_data)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression,Ridge,Lasso
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC,SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB,MultinomialNB
```

```python
original_models = {
    #'Linear_Regression'     : LinearRegression(),
   'Logistic_Regression'   : LogisticRegression(),
    #'Ridge'                 : Ridge(),
   # 'Lasso'                 : Lasso(),
    'SVC'                   : SVC(),
    #'Linear_SVC'            : LinearSVC(),
    'Decision_tree'         : DecisionTreeClassifier(),
    'GNB'                   : GaussianNB(),
    
}
for name,model in original_models.items():
    model.fit(X_train,y_train)
    print(name+'Trainned')
```

```python
for name,model in original_models.items():
    print(name + ' Accuracy : {:2f}'.format(model.score(X_test,y_test)))
```
