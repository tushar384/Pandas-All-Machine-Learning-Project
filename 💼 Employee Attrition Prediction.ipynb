```python
def preprocessing_inputs(df):
    df = df.copy()
    #  Drop single-value columns and id columns
    df = df.drop(['EmployeeCount','EmployeeNumber','MonthlyRate','Over18','StandardHours'],axis = 'columns')
    # Binary-encode binary columns
    df['Gender'] = df['Gender'].replace({'Female' : 1,'Male': 0})
    df['OverTime'] = df['OverTime'].replace({'Yes' : 1,'No': 0})
    # # Ordinal-encode the BusinessTravel column and MaritalStatus
    df['MaritalStatus'] = df['MaritalStatus'].replace({'Married': 0,'Divorced':1,'Single': 2})
    df['BusinessTravel'] =df['BusinessTravel'].replace({'Travel_Frequently' : 0,'Non-Travel': 1,'Travel_Rarely': 2})
    # one-hot encode df
    df = pd.get_dummies(df,drop_first = True)
    # split df into X,y
    y = df['Attrition']
    X = df.drop(['Attrition'],axis = 'columns')
    # train df 
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
    # scaler X,y
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train),index = X_train.index,columns = X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test),index = X_test.index,columns = X_test.columns)
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test = preprocessing_inputs(train_df)
```
