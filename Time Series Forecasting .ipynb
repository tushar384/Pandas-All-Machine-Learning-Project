```python
def preprocessing_inputs(df):
    df = df.copy()
    ## Remove the standard deviation examples
    # df =df.query(Element == 'Temperature change'')
    df = df.query("Element == 'Temperature change'")
    # # Group countries and take their means
    df = df.groupby('Area').mean()
    # 'Area Code''Months Code''Element Code'
    #above columns they are not need for trainding model 
    # here are two - two option 1 st drop ya 2nd loc
    ## Use only time series data
    df = df.loc[:,'Y1961':]
    # there are more time period i take avg time period
    # Get global mean temperature change
    #df = df.mean()
    df = pd.DataFrame(df.mean().reset_index(drop = False))
    # rename columns
    df.columns = ['ds','y']
    # only year iwant 'Y' remove
    # # Fix year column
    df['ds'] = df['ds'].apply(lambda x: x[1:])
   
    return df
```

```python
time_series = preprocessing_inputs(data)
```
