```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Engineering_salary.csv')
```

```python
data
```

```python
def preprocessing_inputs(df):
    df = df.copy()
    # Gender conver into binary
    df['Gender'] = df['Gender'].apply(lambda x :1 if x=='f' else 0)
    # DOB one hot encode
    df['DOB']  = pd.to_datetime(df['DOB'])
    df['DOB_year'] = df['DOB'].apply(lambda x: x.year)
    df['DOB_month'] = df['DOB'].apply(lambda x : x.month)
    df['DOB_day'] = df['DOB'].apply(lambda x : x.day)
    df = df.drop(['DOB'],axis = 'columns')
    # replace -1 into NaN
    df = df.replace(-1,np.NaN)
    # one - hot encode 
    df = pd.get_dummies(df,drop_first = True)
    # fill NaN value in mean
    for column in ['Domain','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']:
        df[column] = df[column].fillna(df[column].mean())
    # split data into X,y
    y = df['Salary']
    X = df.drop(['Salary'],axis = 'columns')
    # test & train data
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocessing_inputs(data)
```
